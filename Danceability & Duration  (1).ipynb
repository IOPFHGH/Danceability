```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\harsh\Downloads\spotify dataset.csv")
df
```

```python
df.isnull().sum()
```

```python
df.track_name=df.track_name.fillna("VACUUM")
df.track_artist=df.track_artist.fillna("VACUUM")
df.track_album_name=df.track_album_name.fillna("VACUUM") 
```

```python
df.isnull().sum()
```

```python
plt.scatter(df.danceability,df.duration_ms )
```

```python
from sklearn.cluster import KMeans
kmeans_model=KMeans(n_clusters=6)
kmeans_model.fit(df[["danceability","duration_ms"]])
```

```python
kmeans_model.cluster_centers_
```

```python
df["cluster_group"]=kmeans_model.labels_
```

```python
x=df.iloc[:,11:]
x
```

```python
df["cluster_group"].value_counts()
```

```python
import seaborn as sns
sns.scatterplot(x="danceability",y="duration_ms",data=df,hue="cluster_group")
```
